Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulated World Bank population data (structure matches real CSV files)
data = {
    'Country Name': ['India', 'China', 'United States', 'Indonesia', 'Brazil'],
    '2015': [1310152403, 1371220000, 320635163, 257563815, 204471769],
    '2016': [1324517249, 1378665000, 323127515, 260580739, 206163057],
    '2017': [1338676785, 1386395000, 325719178, 263991379, 207833823],
    '2018': [1352642280, 1392730000, 327167434, 267670543, 209469333],
    '2019': [1366417754, 1397715000, 328239523, 270625567, 211049527],
}

# Convert to DataFrame
df = pd.DataFrame(data)
df = df.set_index('Country Name')

# Transpose for plotting
df_T = df.transpose()

# Plotting
df_T.plot(kind='bar', figsize=(12, 6))
plt.title('Total Population by Country (2015â€“2019)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: Male and female population for selected countries in 2022
data = {
    'Country': ['India', 'China', 'USA', 'Indonesia', 'Brazil'],
    'Male': [717000000, 742000000, 163000000, 138000000, 105000000],
    'Female': [676000000, 702000000, 169000000, 135000000, 108000000]
}

df = pd.DataFrame(data)
df.set_index('Country', inplace=True)

# Plotting
df.plot(kind='bar', stacked=True, figsize=(10, 6), color=['blue', 'pink'])
plt.title('Gender Distribution by Country (2022)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```
